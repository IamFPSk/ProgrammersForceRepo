# 1. Which month saw the highest joining of the employees

```python
print(df['Month of Joining'].value_counts())
print("This Month saw the highest joining of the employees: ",df['Month Name of Joining'].value_counts().idxmax())
```

# 2. What is the mean salary is the state KS

```python
KS_employees = df[df['State'] == 'KS']
KS_employees['Salary'].mean()
```

# 3. Which Region has highest mean salary

```python
print(df.groupby('Region')['Salary'].mean())
print("This Region has the highest mean Salary:", df.groupby('Region')['Salary'].mean().idxmax())
```

# 4. Which region has lowest mean salary

```python
print("This Region has the highest mean Salary:", df.groupby('Region')['Salary'].mean().idxmin())
```

# 5. Which City has highest mean weight

```python
print(df.groupby('City')["Weight in Kgs."].mean())
print("This city has the highest mean weight: ",df.groupby('City')["Weight in Kgs."].mean().idxmax())
```

# 6. What is the difference between mean salary of Doctors and Professors

```python
print(df.groupby("Name Prefix")["Salary"].mean())
doctor = df[(df['Name Prefix'] == 'Dr.') | (df['Name Prefix'] == 'Drs.')]
professors = df[df['Name Prefix'] == 'Prof.']

DrSalary  = doctor["Salary"].mean()
ProfSalary = professors['Salary'].mean()

salary_difference = DrSalary - ProfSalary

print("The difference between the mean salary of doctors and professors is:", salary_difference)
```

# 7. How many female use yahoo

```python
IdsWithYahoo = df[df["E Mail"].str.contains('yahoo',case=False)]
FWithYahoo = IdsWithYahoo[IdsWithYahoo["Gender"] == "F"]
display(FWithYahoo)
print("Total Female Yahoo users are: ",len(FWithYahoo))
```

# 8. How many Doctors use gmail

```python
IdsWithGmail = df[df["E Mail"].str.contains('gmail',case=False)]
DrsWithGmail = IdsWithGmail[(IdsWithGmail["Name Prefix"] == "Drs.") | (IdsWithGmail["Name Prefix"] == "Dr.")]
display(DrsWithGmail)
print("Total Drs with gmail ids are : ", len(DrsWithGmail))
```

# 9. What is the password of the youngest male

```python
males = df[df['Gender'] == 'M']
youngestMaleIndex = males['Age in Yrs.'].idxmin()
print(df.loc[youngestMaleIndex, "Password"])
```
